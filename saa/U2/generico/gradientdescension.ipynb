```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
X = np.linspace(-5,5,100)
Y = 3 * X**3 - 2 * X + 4

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(X, Y, color="red")
DY = 6 * X**2 - 2
ax.plot(X, DY, color="blue")
```
