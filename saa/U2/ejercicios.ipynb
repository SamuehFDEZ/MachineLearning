## EJERCICIO 10:
En el ejemplo 18 (enlace) se calculan los VIF del ejemplo 17 y se elimina una de las columnas que puede causar colinealidad. Modifica el código para que siga descartando columnas mientras alguna pase el límite aconsejable de VIF. Muestra una captura de su ejecución.<br>

```python
import statsmodels.graphics.api as smg
import pandas as pd
from sklearn import preprocessing
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Cargar datos
df = pd.read_csv('recursos/Housing_Modified.csv')

# Convertir campos binarios a campos booleanos numéricos
lb = preprocessing.LabelBinarizer()
df.driveway = lb.fit_transform(df.driveway)
df.recroom = lb.fit_transform(df.recroom)
df.fullbase = lb.fit_transform(df.fullbase)
df.gashw = lb.fit_transform(df.gashw)
df.airco = lb.fit_transform(df.airco)
df.prefarea = lb.fit_transform(df.prefarea)

# Asegurarse de que las columnas son enteros
df.driveway = df.driveway.astype(int)
df.recroom = df.recroom.astype(int)
df.fullbase = df.fullbase.astype(int)
df.gashw = df.gashw.astype(int)
df.airco = df.airco.astype(int)
df.prefarea = df.prefarea.astype(int)

# Crear variables dummy para stories
df_stories = pd.get_dummies(df['stories'], prefix='stories', drop_first=True)
# Unir las variables dummy al dataframe principal
df = pd.concat([df, df_stories], axis=1)
del df['stories']

# Función para calcular y eliminar variables con VIF alto
def eliminar_colinealidad(df, umbral=10):
    while True:
        # Filtrar solo columnas numéricas
        df_numeric = df.select_dtypes(include=[float, int])

        vif_datos = pd.DataFrame()
        vif_datos["feature"] = df_numeric.columns
        vif_datos["VIF"] = [variance_inflation_factor(df_numeric.values, i) for i in range(len(df_numeric.columns))]

        print(vif_datos)  # Imprimir VIF actual

        # Identificar la variable con VIF más alto
        max_vif = vif_datos.loc[vif_datos["VIF"].idxmax()]

        # Si el VIF máximo es menor al umbral, detener el proceso
        if max_vif["VIF"] < umbral:
            break

        # Eliminar la variable con mayor VIF
        print(f"Eliminando '{max_vif['feature']}' con VIF de {max_vif['VIF']:.2f}")
        df.drop(columns=[max_vif["feature"]], inplace=True)

    return df

# Aplicar eliminación iterativa de colinealidad
df = eliminar_colinealidad(df)

# Dibujar la matriz de correlaciones usando plot_corr() de Statsmodels
corr = df.corr()  # Crea matriz de correlaciones
smg.plot_corr(corr, xnames=list(corr.columns))
plt.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.graphics.api as smg
from statsmodels.graphics.regressionplots import plot_leverage_resid2
import matplotlib.pyplot as plt
from scipy import stats

# Ajustar un modelo de regresión, por ejemplo
X = df.drop(columns=['price'])
X = sm.add_constant(X)  # Agregar la constante
y = df['price']

modelo = sm.OLS(y, X).fit()

# 1. Graficar los residuos normalizados contra el leverage
fig, ax = plt.subplots(figsize=(8, 6))
fig = plot_leverage_resid2(modelo, ax=ax)
plt.show()

# 2. Test de outliers de Bonferroni
test = modelo.outlier_test()
print(test['bonf(p)'])
print('Puntos de datos malos (bonf(p) < 0.05):', test[test['bonf(p)'] < 0.05])

# 3. Detección de outliers usando Z-scores
z_scores = np.abs(stats.zscore(X.drop('const', axis=1)))  # Excluimos la constante
outliers_z = np.where(z_scores > 3)  # Detectamos outliers con Z-scores mayores a 3
print(f"Outliers según Z-scores: {outliers_z}")

# 4. Detección de outliers usando IQR (Interquartile Range)
Q1 = X.drop('const', axis=1).quantile(0.25)
Q3 = X.drop('const', axis=1).quantile(0.75)
IQR = Q3 - Q1

# Identificar los outliers
outliers_iqr = ((X.drop('const', axis=1) < (Q1 - 1.5 * IQR)) | (X.drop('const', axis=1) > (Q3 + 1.5 * IQR)))
print(f"Outliers según IQR:\n{outliers_iqr}")

# Puedes visualizar si coinciden con los valores obtenidos por el test de Bonferroni
```
