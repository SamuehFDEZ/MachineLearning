## EJERCICIO 10:
En el ejemplo 18 (enlace) se calculan los VIF del ejemplo 17 y se elimina una de las columnas que puede causar colinealidad. Modifica el código para que siga descartando columnas mientras alguna pase el límite aconsejable de VIF. Muestra una captura de su ejecución.<br>

```python
import statsmodels.graphics.api as smg
import pandas as pd
from sklearn import preprocessing
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor

df = pd.read_csv('recursos/Housing_Modified.csv')

# Convertir campos binarios a campos booleanos numéricos
lb = preprocessing.LabelBinarizer()
df.driveway = lb.fit_transform(df.driveway)
df.recroom = lb.fit_transform(df.recroom)
df.fullbase = lb.fit_transform(df.fullbase)
df.gashw = lb.fit_transform(df.gashw)
df.airco = lb.fit_transform(df.airco)
df.prefarea = lb.fit_transform(df.prefarea)

# Asegurarse de que las columnas son enteros
df.driveway = df.driveway.astype(int)
df.recroom = df.recroom.astype(int)
df.fullbase = df.fullbase.astype(int)
df.gashw = df.gashw.astype(int)
df.airco = df.airco.astype(int)
df.prefarea = df.prefarea.astype(int)

# Crear variables dummy para stories
df_stories = pd.get_dummies(df['stories'], prefix='stories', drop_first=True)
# Unir las variables dummy al dataframe principal
df = pd.concat([df, df_stories], axis=1)
del df['stories']

# Función para calcular y eliminar variables con VIF alto
def eliminar_colinealidad(df, umbral=10):
    while True:
        # Filtrar solo columnas numéricas
        df_numeric = df.select_dtypes(include=[float, int])

        vif_datos = pd.DataFrame()
        vif_datos["feature"] = df_numeric.columns
        vif_datos["VIF"] = [variance_inflation_factor(df_numeric.values, i) for i in range(len(df_numeric.columns))]

        print(vif_datos)  # Imprimir VIF actual

        # Identificar la variable con VIF más alto
        max_vif = vif_datos.loc[vif_datos["VIF"].idxmax()]

        # Si el VIF máximo es menor al umbral, detener el proceso
        if max_vif["VIF"] < umbral or max:
            break

        # Eliminar la variable con mayor VIF
        print(f"Eliminando '{max_vif['feature']}' con VIF de {max_vif['VIF']:.2f}")
        df.drop(columns=[max_vif["feature"]], inplace=True)

    return df

# Aplicar eliminación iterativa de colinealidad
df = eliminar_colinealidad(df)

# Dibujar la matriz de correlaciones usando plot_corr() de Statsmodels
corr = df.corr()  # Crea matriz de correlaciones
smg.plot_corr(corr, xnames=list(corr.columns))
plt.show()
```

## EJERCICIO 11.
Ejecuta el ejemplo 21 y el ejemplo 22 aplicado al modelo del ejercicio anterior. Indica qué
datos son outliers. Ahora aplica alguno de los algoritmos básicos (o el basado en z-scores o el basado en
IQR) y comprueba si detectan los mismos outliers.<br>

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.graphics.api as smg
from statsmodels.graphics.regressionplots import plot_leverage_resid2
import matplotlib.pyplot as plt
from scipy import stats

df = pd.read_csv('recursos/Housing_Modified.csv')

# Seleccionar solo las columnas numéricas
df = df.select_dtypes(include=[np.number])

# Rellenar valores NaN con la media de las columnas
df.fillna(df.mean(), inplace=True)

# Definir las variables dependientes e independientes
X = df.drop(columns=['price'])
X = sm.add_constant(X)  # Agregar la constante
y = df['price']

# Eliminar filas con valores infinitos
X = X.replace([np.inf, -np.inf], np.nan).dropna()
y = y[X.index]  # Asegurarse de que 'y' tenga las mismas filas que 'X'

# Ajustar el modelo de regresión
modelo = sm.OLS(y, X).fit()

# 1. Graficar los residuos normalizados contra el leverage
fig, ax = plt.subplots(figsize=(8, 6))
fig = plot_leverage_resid2(modelo, ax=ax)
plt.show()

# 2. Test de outliers de Bonferroni
test = modelo.outlier_test()
print(test['bonf(p)'])
print('Puntos de datos malos (bonf(p) < 0.05):', test[test['bonf(p)'] < 0.05])

# 3. Detección de outliers usando Z-scores
z_scores = np.abs(stats.zscore(X.drop('const', axis=1)))  # Excluimos la constante
outliers_z = np.where(z_scores > 3)  # Detectamos outliers con Z-scores mayores a 3
print(f"Outliers según Z-scores: {outliers_z}")

# 4. Detección de outliers usando IQR (Interquartile Range)
Q1 = X.drop('const', axis=1).quantile(0.25)
Q3 = X.drop('const', axis=1).quantile(0.75)
IQR = Q3 - Q1

# Identificar los outliers
outliers_iqr = ((X.drop('const', axis=1) < (Q1 - 1.5 * IQR)) | (X.drop('const', axis=1) > (Q3 + 1.5 * IQR)))
print(f"Outliers según IQR:\n{outliers_iqr}")
```

## EJERCICIO 12:
Modifica el ejemplo 28 para calcular también un modelo lineal a los mismos datos y
muestra el R2 del modelo lineal y del modelo polinómico con los datos de entrenamiento<br>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Número de datos
n = 100
np.random.seed(0)  # Hacerlo repetible

# Generación de datos de entrada (x) y salida (y)
x = np.sort(10 * np.random.rand(n, 1) - 5, axis=0)
y = 0.2 * x**3 - 0.5 * x + 2 + np.random.randn(n, 1)  # Relación no lineal con ruido

# Crear modelo lineal
rl = LinearRegression().fit(x, y)

# Crear modelo polinómico (añadiendo x**3 como característica)
x_poli = np.hstack([x, x**3])
rp = LinearRegression().fit(x_poli, y)

# Crear datos de prueba para dibujar las predicciones
rejilla_lineal = np.linspace(-5, 5, 100).reshape(-1, 1)
rejilla_polino = np.hstack([rejilla_lineal, rejilla_lineal**3])

# Graficar los datos y las predicciones de ambos modelos
plt.scatter(x, y, color="black", label="Datos")
plt.plot(rejilla_lineal, rl.predict(rejilla_lineal), color="blue", label="Lineal")
plt.plot(rejilla_lineal, rp.predict(rejilla_polino), color="red", label="Polinómica")
plt.legend()
plt.show()

# Calcular R² para el modelo lineal con los datos de entrenamiento
r2_lineal = rl.score(x, y)
# Calcular R² para el modelo polinómico con los datos de entrenamiento
r2_polinomico = rp.score(x_poli, y)

# Mostrar los resultados
print(f"R² del modelo lineal: {r2_lineal:.4f}")
print(f"R² del modelo polinómico: {r2_polinomico:.4f}")
```

## EJERCICIO 15:
Calcula la precisión y el recall del ejemplo de la matriz de confusión de la página 71. <br>


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import confusion_matrix, precision_score, recall_score

# Generación de datos
n = 100  # nº de datos
np.random.seed(0)  # Hacerlo repetible
X = 6 * np.random.rand(n, 1) - 3
X = np.sort(X, axis=0)  # Los ordeno para dibujarlos bien
y = 0.5 * X**2 + X + 2 + np.random.randn(n, 1)  # Relación no lineal con ruido

# Generar datos de potencias de los originales
x_polinomial = PolynomialFeatures(degree=2, include_bias=False)
X_polinomial = x_polinomial.fit_transform(X)

# Aplicar el modelo
rl = LinearRegression()
rl.fit(X_polinomial, y)

# Predicciones
y_pred = rl.predict(X_polinomial)

# Convertir a categorías (por ejemplo, valores mayores que la media de y como "1" y menores que la media como "0")
y_mean = np.mean(y)
y_true_class = (y > y_mean).astype(int)
y_pred_class = (y_pred > y_mean).astype(int)

# Calcular la matriz de confusión
cm = confusion_matrix(y_true_class, y_pred_class)
print("Matriz de Confusión:")
print(cm)

# Calcular precisión y recall
precision = precision_score(y_true_class, y_pred_class)
recall = recall_score(y_true_class, y_pred_class)

print(f"Precisión: {precision:.4f}")
print(f"Recall: {recall:.4f}")

# Dibujar el modelo de regresión polinómica
titulo = f"Regresión Polinómica con R2: {rl.score(X_polinomial, y)}"
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(X, rl.predict(X_polinomial), color='red', linewidth=2)
plt.title(titulo)
plt.show()
```

```python
from sklearn.ensemble import AdaBoostClassifier, VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
import pandas as pd

# Leer los datos
df = pd.read_csv("recursos/Diabetes.csv")
X = df[['age', 'serum_insulin']]  # Variables independientes
y = df['class'].values  # Variables dependientes
X = StandardScaler().fit_transform(X)  # Normalizar los datos

# Evaluar el modelo
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2017)
kf = StratifiedKFold(n_splits=5, random_state=2017, shuffle=True)
n_arboles = 100

# Árbol con 5-fold CV y restricción de max_depth a 1 para tener impurezas
dt = DecisionTreeClassifier(max_depth=1, random_state=2017)

# AdaBoost con Decision Tree como base
dt_b = AdaBoostClassifier(estimator=dt, n_estimators=n_arboles, learning_rate=0.1, algorithm='SAMME', random_state=2017)

# Crear el meta-modelo con VotingClassifier (Hard Voting)
voting_clf = VotingClassifier(estimators=[('dt', dt), ('dt_b', dt_b)], voting='hard')

# Ajustar el modelo VotingClassifier en los datos de entrenamiento
voting_clf.fit(X_train, y_train)

# Evaluar el meta-modelo con cross-validation
r = cross_val_score(voting_clf, X_train, y_train, cv=kf)
print("Voting Classifier (Hard Voting) - Train Accuracy:", r.mean())

# Evaluar la precisión en el conjunto de prueba
y_pred = voting_clf.predict(X_test)
print("Voting Classifier (Hard Voting) - Test Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

d) Modifica el meta-modelo soft-voting pasándole esos 5 modelos base, y además le pasas los
pesos modificados de manera que el peso del modelo i-ésimo sea su accuracy dividido por la
suma de todas las accuracies.<br>

```python
from sklearn.ensemble import AdaBoostClassifier, VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
import pandas as pd

# Leer los datos
df = pd.read_csv("recursos/Diabetes.csv")
X = df[['age', 'serum_insulin']]  # Variables independientes
y = df['class'].values  # Variables dependientes
X = StandardScaler().fit_transform(X)  # Normalizar los datos

# Evaluar el modelo
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2017)
kf = StratifiedKFold(n_splits=5, random_state=2017, shuffle=True)
n_arboles = 100

# Árbol con 5-fold CV y restricción de max_depth a 1 para tener impurezas
dt = DecisionTreeClassifier(max_depth=1, random_state=2017)

# AdaBoost con Decision Tree como base
dt_b = AdaBoostClassifier(estimator=dt, n_estimators=n_arboles, learning_rate=0.1, algorithm='SAMME', random_state=2017)

# Lista de modelos base
models = [
    ('dt', dt),
    ('dt_b', dt_b),
]

# Calcular las accuracies de cada modelo base
accuracies = []
for name, model in models:
    model.fit(X_train, y_train)
    accuracy = metrics.accuracy_score(y_train, model.predict(X_train))
    accuracies.append(accuracy)

# Calcular los pesos basados en las accuracies
total_accuracy = sum(accuracies)
weights = [accuracy / total_accuracy for accuracy in accuracies]

# Crear el meta-modelo con VotingClassifier (Soft Voting) y los pesos
voting_clf = VotingClassifier(estimators=models, voting='soft', weights=weights)

# Ajustar el modelo VotingClassifier en los datos de entrenamiento
voting_clf.fit(X_train, y_train)

# Evaluar el meta-modelo con cross-validation
r = cross_val_score(voting_clf, X_train, y_train, cv=kf)
print("Voting Classifier (Soft Voting) - Train Accuracy:", r.mean())

# Evaluar la precisión en el conjunto de prueba
y_pred = voting_clf.predict(X_test)
print("Voting Classifier (Soft Voting) - Test Accuracy:", metrics.accuracy_score(y_test, y_pred))
```
